```python
from transformers import pipeline

# Initialize a sentiment analysis pipeline
classifier = pipeline('sentiment-analysis',model="facebook/bart-large-mnli")

# Define the list of sentences
sentences = [
    "Once upon a time, I lived in my village ",
    "In a world where lots of people are weak",
    "The future of technology is unimaginable",
    "The team won the championship",
    "A mysterious figure appeared informed of me ",
    "On a sunny day , I don't like to go out ",
    "A new discovery was made by me ",
    "The team celebrated their victory in the ground ",
    "The ancient ruins were found beside the garden ",
    "A new adventure began today ",
    "The book was a best-seller of all time "
]

# Perform sentiment analysis
results = classifier(sentences)

# Print each result on a new line
for i, (sentence, result) in enumerate(zip(sentences, results)):
    print(f"Sentence {i + 1}: {sentence}")
    print(f"Sentiment: {result['label']} (Score: {result['score']:.2f})\n")
```

```python
from transformers import AutoTokenizer

# load a pre trained Model :
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

#text : which we want to tokenize :-
text = " I was a boy , who was afraid of large concepts "

# Tokenize the text ;
tokens = tokenizer.tokenize(text)

# Print the tokens :
# returns a list of tokens
print("Tokens :" ,tokens)

# convert tokens to token-id :
input_id = tokenizer.convert_tokens_to_ids(tokens)

# Print the input ids :
print("Input-ids :" , input_id)

# encode the text : {Tokenization + converting to input Ids }
encode_input = tokenizer(text)

# Print the encode input :
print("Encode Input :" , encode_input)

# decode the output :

decode_input = tokenizer.decode(input_id)

# Print the decode input :
print("Decode Input :" , decode_input)

print(tokenizer.decode(encode_input["input_ids"]))
```
